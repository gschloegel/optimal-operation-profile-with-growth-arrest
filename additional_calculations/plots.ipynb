In this notebook the schematic representation of the fed-batch processes with and without feed are created.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', size=14)
```

```python
t0 = np.linspace(0, 10, 50)
t1 = np.linspace(10, 20, 50)
t2 = np.linspace(20, 30, 50)
t = np.concatenate((t0, t1, t2))
```

```python
# biomass

x0 = 5 / np.exp(5) * np.exp(0.5 * t0)
x1 = t1 - 5
x2 = t2 * 0 + 15
```

```python
# feed
f0 = t0 * 0
f1 = t1 * 0 + 1
f2 = t2 * 0 + 1
f = np.concatenate([f0, f1, f2])
```

```python
# glucose concentration
g0 = 1 - np.exp(-5) * np.exp(0.5 * t0)
g1 = t1 * 0
g2 = t2 * 0
g = np.concatenate((g0, g1, g2))
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=((15,5)))

ax0.plot(t, f, '-', label='Glucose fed')
ax0.plot(t, g, '--', label='Glucose concentration')

ax0.set_xticks((0, 10, 20), labels=("Start batch", "End batch", "Start starvation"))
ax0.set_yticks([0])
ax0.set_xlabel("Time [h]")
ax0.set_ylabel("$f_{feed}$")
# ax0.legend()

ax0.axvline(x=10, ls="dotted", c='k')
ax0.axvline(x=20, ls="dotted", c='k')

ax0a = ax0.twinx()
ax0a.set_ylabel("Glucose feed $[g h^{-1}]$")
ax0.set_ylabel("Glucose concentration $[gL^{-1}]$")
ax0a.yaxis.label.set_color('blue')
ax0.yaxis.label.set_color('orange')
ax0a.set_yticks([])


ax1.plot(t0, x0, '-', label='Batch phase (exponential growth)')
ax1.plot(t1, x1, '--', label='Growth phase (linear growth)')
ax1.plot(t2, x2, '-.', label='Starvation phase (no growth)')

ax1.set_xticks((0, 10, 20), labels=("Start batch", "End batch", "Start starvation"))
ax1.set_yticks([0])
ax1.set_xlabel("Time [h]")
ax1.set_ylabel("Biomass [g DW]")
ax1.legend()


ax1.axvline(x=10, ls="dotted", c='k')
ax1.axvline(x=20, ls="dotted", c='k')

fig.tight_layout()

plt.savefig("../plots/phases_with_starvation.png", facecolor="white", bbox_inches="tight")
plt.show()
```

```python
# Plot the process without starvation

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=((15,5)))

t = np.concatenate([t0, t1,])
f = np.concatenate([f0, f1,])
g = np.concatenate([g0, g1,])

ax0.plot(t, f, '-', label='Glucose fed')
ax0.plot(t, g, '--', label='Glucose concentration')

ax0.set_xticks((0, 10), labels=("Start batch", "End batch"))
ax0.set_yticks([0])
ax0.set_xlabel("Time $[h]$")
# ax0.set_ylabel("")
# ax0.legend()

ax0.axvline(x=10, ls="dotted", c='k')
ax0.axvline(x=20, ls="dotted", c='k')

ax0a = ax0.twinx()
ax0a.set_ylabel("Glucose feed $[g h^{-1}]$")
ax0.set_ylabel("Glucose concentration $[gL^{-1}]$")
ax0a.yaxis.label.set_color('blue')
ax0.yaxis.label.set_color('orange')
ax0a.set_yticks([])


ax1.plot(t0, x0, '-', label='Batch phase (exponential growth)')
ax1.plot(t1, x1, '--', label='Growth phase (linear growth)')

ax1.set_xticks((0, 10), labels=("Start batch", "End batch"))
ax1.set_yticks([0])
ax1.set_xlabel("Time $[h]$")
ax1.set_ylabel("Biomass $[g \\ DW]$")
ax1.legend()


ax1.axvline(x=10, ls="dotted", c='k')
ax1.axvline(x=20, ls="dotted", c='k')

fig.tight_layout()

plt.savefig("../plots/phases", facecolor="white", bbox_inches="tight")
plt.show()
```
